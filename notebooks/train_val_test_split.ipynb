```python
clean_df = pd.read_parquet(Path.cwd().parent / 'data' / 'clean.parquet')
clean_df.head()
```

```python
TRAIN_END = 20
VAL_END = 25
DATASET_LENGTH = len(clean_df)
OVERALL_SAR_FRAC = len(clean_df[clean_df['isSAR'] == 1]) / DATASET_LENGTH

train_size = len(clean_df[clean_df['day'] < TRAIN_END])
val_size = len(
    clean_df[(clean_df['day'] >= TRAIN_END) & (clean_df['day'] < VAL_END)]
    )
test_size = len(clean_df[clean_df['day'] >= VAL_END])

train_pct = train_size / DATASET_LENGTH
val_pct = val_size / DATASET_LENGTH
test_pct = test_size / DATASET_LENGTH

train_df = clean_df[clean_df['day'] < TRAIN_END]
val_df = clean_df[(clean_df['day'] >= TRAIN_END) & (clean_df['day'] < VAL_END)]
test_df = clean_df[clean_df['day'] >= VAL_END]

train_target_frac = len(train_df[train_df['isSAR'] == 1]) / len(train_df)
val_target_frac = len(val_df[val_df['isSAR'] == 1]) / len(val_df)
test_target_frac = len(test_df[test_df['isSAR'] == 1]) / len(test_df)

print(f'Sizes: train {train_size} val {val_size} test {test_size}')
print(f'Percent of original rows: train {train_pct:.2%} val {val_pct:.2%} test {test_pct:.2%}')
print(f'Overall SAR percentage: {OVERALL_SAR_FRAC:.2%}')
print(f'SAR percents: train {train_target_frac:.4%} val {val_target_frac:.4%} test {test_target_frac:.4%}')
```
